```python
def add_book():
    while 1: 
        title_book=input('Enter the title of the book:') 
        author_book=input('Enter author of book:') 
        quantity_book = int(input('Enter the quantity of book:')) 
        price_book = int(input('Enter the price of book')) 
        num1=input('To stop the loop...write stop, otherwise ignore it:\n')
        if num1.lower() =='stop':
            print('loop stopped') 
            break
        elif (num1!='stop'):
            continue
     
        else: 
            print(f'you enetered {num1}')
    key_value = {
            'Title of book': title_book, 
            'Author of book': author_book,
            'Quantity': quantity_book,
            'Price': price_book
        }
        
    dic['Added Books'].append(key_value) 
    print(dic)
    return dic
# x-----------------------x----------------------------------------------------------x

def del_book():
  del1= int(input("To remove any book enter ----> 1 --- &----- to remove whole data enter -----> 2:"))
  print('You entered :' ,del1) 
  book=int(input('Enter the book number to delete :'))
  remove_books = dic['Books']
  new_remove = dic['Added Books']
  if del1== 1:
      # book==int(input('Enter the number of book to delete it:'))
      if book==1:
          remove_books.pop(0)
      elif book==2:
          remove_books.pop(1)
      elif book==3:
          remove_books.pop(2)
      elif book==4:
         remove_books.pop(3)
      elif book==5:
          remove_books.pop(4)
      elif book==6:
          remove_books.pop(5)
      elif book==7:
           remove_books.pop(6)
      elif book==8:
           remove_books.pop(7)
      elif book==9:
           remove_books.pop(8)
      elif book==10:
           remove_books.pop(9)     
      elif book==11:
           remove_books.pop(10)
      elif book==12:
          remove_books.pop(11)           
  elif del1 == 2:
      dic.clear()
      return dic
  else:
      print('Enter a valid number!')
  print("After Deletion ....\n Updated Data;" ,dic)
  return dic
   
# x--------------------------x--------------------------------------Lx
def sale_transaction():
    num11=int(input("Kindly enter the book's indexing: "))
    if num11>0:
        data = dic['Books']
        piece= int(input(f"Ready to add to your cart? How many copies of {data[num11]['Title']} would you like? :"))
        print(data[num11])
        multiply= piece * data[num11]['Price']
        print(f"""The price of one book is {data[num11]['Price']} $\nSo, the price of {piece} copies of {data[num11]['Title']} is {multiply} $""")
        
# print({data[num11]['Price']})       
        
# x--------------------------x--------------------------------------Lx

def total_cost():    
    book = dic['Books']
    total_quantity=0 
    total_price = 0
    for j in book:
        total_price+=j['Price']*j['Quantity']
        total_quantity+= j['Quantity']
    print('The total quantity of books is---->',total_quantity)
    print(f"The Total Price of inventory is,----> {total_price} $")
# x--------------------------x--------------------------------------Lx

def greet():
    name = input("Enter your name:")
    option = (f"""Welcome to my store: {name}!
    Choose option: 
    1----> view store 
    2--->Add item
    3 ---->delete item
    4-----> To make purchase""")
    print( option)
    num = int(input('Enter a number:'))
    count=1
    if (num == 1):
        for key,value in dic.items():
            print(f" \n     Category  :   {key}")
            for items in value:
                print(f"book indexing  {count} : {items}")
                count+=1
        print('This was the data in the Inventory!\n\n')
                
        print(total_cost())              
    elif (num ==2):
        add_book() 
    elif(num==3):
        for key,value in dic.items():
          print(f" \n     Category  :   {key}")
          for items in value:
              print(f"  {count} : {items}")
              count+=1
        del_book()
    elif (num==4):
        for key,value in dic.items():
            print(f" \n     Category  :   {key}")
            for items in value:
                print(f"book indexing  {count} : {items}")
                count+=1
        print('This was the data in the Inventory!\n\n')
                
        print(sale_transaction())


print(greet())


```
